```python
import pandas as pd
```

```python
df = pd.read_excel(r"/content/Customer Call List.xlsx")
```

```python
df
```

```python
#remove duplicate records
df.drop_duplicates()
```

```python
df #save the non-duplicate data in df
```

```python
#remove unusable columns
df.drop(columns = 'Not_Useful_Column')
df
```

```python
#strip takes value from both sides - lstrip and rstrip would be outside left and right respectively
#by default it just takes whitespace
df['Last_Name'].str.strip()
```

```python
#removing the dots in last name
df['Last_Name'].str.lstrip("...")
```

```python
#but we want to save it!
df['Last_Name'] = df['Last_Name'].str.lstrip("...")
df['Last_Name'] = df['Last_Name'].str.lstrip("/")
df['Last_Name'] = df['Last_Name'].str.rstrip("_")
```

```python
df
```

```python
#reset the entire progress and lets do this another way
df['Last_Name'] = df['Last_Name'].str.strip("123._/")
```

```python
df
```

```python
#now phone numbers
#first strip it down just the numbers
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
```

```python
df
```

```python
#first convert the phone numbers into string to manipulate the format
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
```

```python
#placing - in between the numbers
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python
#now remove the Nan-- and Na--
df["Phone_Number"] = df["Phone_Number"].str.replace("nan--", '')
df["Phone_Number"] = df["Phone_Number"].str.replace("Na--", '')
```

```python
df
```

```python
#address cleaning
df[["Street_Address", "State", "Zip_code"]] = df['Address'].str.split(',', n = 2, expand = True) #looking at the very first and second commas
```

```python
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')
```

```python
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N')
```

```python
df
```

```python
df = df.replace("N/a", '')
df = df.replace("NaN", '')
```

```python
df
```

```python
df = df.fillna('')
df
```

```python
for x in df.index:
  if df.loc[x, "Do_Not_Contact"] == 'Y':
    df.drop(x, inplace = True)
```
